```python
# Imports pandas but changes reference to it to be 'pd' so it can be called with 'pd' later
import pandas as pd
```

```python
%matplotlib inline
```

```python
# Import in the CSV file as prop_list variable
prop_list = pd.read_csv("http://www.firstpythonnotebook.org/_static/committees.csv")
```

```python
# Import in the CSV file as contrib_list variable
contrib_list = pd.read_csv("http://www.firstpythonnotebook.org/_static/contributions.csv")
```

```python
# Count the number of times each unique name appears in the prop_name column
prop_list.prop_name.value_counts()
```

```python
# Filter out the list where prop_name references marijuana, and assign it to variable 'prop'
prop = prop_list[prop_list.prop_name == 'PROPOSITION 064- MARIJUANA LEGALIZATION. INITIATIVE STATUTE.']
```

```python
# A 'wildcard' filter for the same thing...
prop_list[prop_list.prop_name.str.contains("MARIJUANA")]
```

```python
prop.info()
```

```python
prop.head(11)
```

```python
prop.committee_position.value_counts()
```

```python
merged = pd.merge(prop, contrib_list, on="calaccess_committee_id")
```

```python
merged.info()
```

```python
merged.amount.sum()
```

```python
# Counting number of contributions in support/opposition
merged.committee_position.value_counts()
```

```python
# Filter out fields where committee_position column has value 'SUPPORT', and assign it to variable 'support'
support = merged[merged.committee_position == 'SUPPORT']
```

```python
# Filter out fields where committee_position column has value 'OPPOSE', and assign it to variable 'oppose'
oppose = merged[merged.committee_position == 'OPPOSE']
```

```python
# Total contributions from amount column for support
support.amount.sum()
```

```python
# Total contributions from amount column for oppose
oppose.amount.sum()
```

```python
# Calculate the percentage of 'support' contributions from the total (merged table)
support.amount.sum() / merged.amount.sum()
```

```python
# Sort the values in support variable, DESCENDING
support.sort_values("amount", ascending=False)
```

```python
# Count only the number of contributions that are above 0
merged[merged.amount > 0].committee_position.value_counts()
```

```python
oppose.sort_values("amount", ascending=False)
```

```python
bigdogs = merged.groupby(
    ['contributor_firstname', 'contributor_lastname', 'committee_position']
).amount.sum().reset_index().sort_values("amount", ascending=False)
```

```python
bigdogs
```

```python
bigdogs[bigdogs.amount >= 100000].amount.sum() / merged.amount.sum()
```

```python
merged.groupby(
    "contributor_state"
).amount.sum().reset_index().sort_values(
    "amount",
    ascending=False
)
```

```python
bigdogs.to_csv("bigdogs.csv", index=False)
```

```python
merged[ 'in_state' ] = merged.contributor_state == 'CA'
```

```python
merged.groupby(
    "in_state"
).amount.sum().reset_index().sort_values(
    "amount",
    ascending=False
)
```

```python
# Replaces null values with an empty string
# Could also use merged.contributor_firstname = merged.contributor_firstname.fillna("")
# OR... merged.contributor_firstname.fillna("", inplace=True)
merged[ 'clean_firstname' ] = merged.contributor_firstname.fillna("")
```

```python
merged[ 'clean_fullname'] = merged.clean_firstname + " " + merged.contributor_lastname
```

```python
merged.head()
```

```python
bigdogs = merged.groupby(
    ['clean_fullname', 'committee_position']
).amount.sum().reset_index().sort_values("amount", ascending=False)
```

```python
bigdogs.head(6)
```

```python
def standardise_name(clean_fullname):
    clean_fullname = clean_fullname.strip()
    if clean_fullname == "FUND FOR POLICY REFORM (NONPROFIT 501 (C)(4))":
        return "FUND FOR POLICY REFORM"
    elif clean_fullname == "SEAN PARKER AND AFFILIATED ENTITIES":
        return "SEAN PARKER"
    else: return clean_fullname
```

```python
merged
```

```python
merged['standardised_name' ] = merged.clean_fullname.apply(standardise_name)
```

```python
merged
```

```python
bigdogs = merged.groupby(
    ['standardised_name', 'committee_position']
).amount.sum().reset_index().sort_values("amount", ascending=False)
```
